```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('cars_final.csv')
print(df.head())
```

```python
df.count()
```

```python
df['New Engine'] = df['Engine'].apply(lambda x: round(float(x)/1000, 1) if not pd.isna(x) else np.nan)
print(df['New Engine'])
```

```python
df['New Price'] = df['Price'].apply(lambda x: round(float(x)/100000) if not pd.isna(x) else np.nan)
print(df['New Price'])
```

```python
df['New Km'] = df['Kilometer Run'].apply(lambda x: round(float(x)/1000) if not pd.isna(x) else np.nan)
print(df['New Km'])
```

```python
from sklearn import linear_model
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Engine'], df['Mileage'], color='red', marker='+')
plt.xlabel('Engine (in CC)')
plt.ylabel('Mileage (in km)')
plt.show()
```

```python
df1 = df[['Engine', 'Mileage']]
df1 = df1.dropna()
print(df1)
```

```python
reg = linear_model.LinearRegression()
#reg.fit(df['Engine'], df['Mileage'])
reg.fit(df1[['Engine']], df1[['Mileage']])
```

```python
#linear regression
plt.figure(figsize=(10,6))
plt.scatter(df1['Engine'], df1['Mileage'])
plt.plot(df1['Engine'], reg.predict(df1[['Engine']]), color='green')
plt.xlabel('Engine (cc)')
plt.ylabel('Mileage (km)')
plt.show()
```

```python
df2 = df[['Make Year', 'Price']]
df2 = df2.dropna()
```

```python
reg2 = linear_model.LinearRegression()
reg.fit(df2[['Make Year']], df2[['Price']])
```

```python
#linear regression
plt.figure(figsize=(10,6))
plt.scatter(df2['Make Year'], df2['Price'])
plt.plot(df2['Make Year'], reg.predict(df2[['Make Year']]), color='red')
plt.xlabel('Make Year')
plt.ylabel('Price')
plt.show()
```
