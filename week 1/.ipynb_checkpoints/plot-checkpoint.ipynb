```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('cars_final.csv')
print(df.head())
```

```python
df.count()
```

```python
df['New Engine'] = df['Engine'].apply(lambda x: round(float(x)/1000, 1) if not pd.isna(x) else np.nan)
print(df['New Engine'])
```

```python
df['New Price'] = df['Price'].apply(lambda x: round(float(x)/100000) if not pd.isna(x) else np.nan)
print(df['New Price'])
```

```python
df['New Km'] = df['Kilometer Run'].apply(lambda x: round(float(x)/1000) if not pd.isna(x) else np.nan)
print(df['New Km'])
```

```python
print(set(df['Kilometer Run'].unique()))
```

```python
df['Brand'].unique()
```

```python
plt.figure(figsize=(10,7))
plt.hist(df['Brand'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,12))

#plot 1
plt.subplot(3,2,1)
plt.plot(df['Mileage'], 'o')
plt.title('Mileage (in km)')

#plot 2
plt.subplot(3,2,2)
plt.plot(df['Make Year'], 'o')
plt.title('Make Year (in AD)')

#plot 3
plt.subplot(3,2,3)
plt.plot(df['New Engine'], 'o')
plt.title('Engine (in litres)')

#plot 4
plt.subplot(3,2,4)
plt.plot(df['New Price'], 'o')
plt.title('Price (in lakhs)')

#plot 5
plt.subplot(3,2,5)
plt.plot(df['New Km'], 'o')
plt.title('Kilometer Run (in thousands)')

plt.tight_layout(rect=[0,0.03,1,0.95])
plt.suptitle('Distribution of Car Data')
plt.show()
```

```python
fig, ax = plt.subplots(3,2, figsize=(8,10))

#plot 1
ax[0,0].hist(df['Mileage'])
ax[0,0].set_title('Mileage (in km)')

#plot 2
ax[0,1].hist(df['Make Year'])
ax[0,1].set_title('Make Year (in AD)')

#plot 3
ax[1,0].hist(df['New Engine'])
ax[1,0].set_title('Engine (in litres)')

#plot 4
ax[1,1].hist(df['New Price'])
ax[1,1].set_title('Price (in lakhs)')

#plot 5
ax[2,0].hist(df['New Km'])
ax[2,0].set_title('Kilometer Run (in thousands)')

fig.suptitle('Histogram of car data')
fig.tight_layout(rect=[0,0.03,1,0.95])
plt.show()
```
