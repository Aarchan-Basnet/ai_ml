```python
#convert age in days to years
#df['age'] = df['age'].apply(lambda x: x//365)
df['age'] = df['age'] // 365.25
```

```python
#calculate BMI as well as bmi categories
df['BMI'] = df['weight'] / (df['height']/100)**2

def categorize_bmi(bmi):
    if bmi < 18.5:
        return 'underweight'
    elif 18.5<=bmi<25:
        return "normalweight"
    elif 25<=bmi<30:
        return 'overweight'
    else:
        return 'obesity'

df['bmi_catz'] = df['BMI'].apply(categorize_bmi)
    
print(df.head())
```

```python

def categorize_blood_pressure(row):
    ap_hi = row['ap_hi']
    ap_lo = row['ap_lo']
    if ap_hi < 90 and ap_lo < 60:
        return "Low blood pressure"
    elif 90 <= ap_hi < 120 and 60 <= ap_lo < 80:
        return "Normal blood pressure"
    elif 120 <= ap_hi < 140 or 80 <= ap_lo < 90:
        return "Prehypertension"
    elif 140 <= ap_hi < 160 or 90 <= ap_lo < 100:
        return "Stage 1 hypertension"
    elif ap_hi >= 160 or ap_lo >= 100:
        return "Stage 2 hypertension"
    else:
        return None
    
df['blood_pressure'] = df.apply(categorize_blood_pressure, axis=1)
```

```python
label_encoder = LabelEncoder()
df['bmi_enc'] = label_encoder.fit_transform(df['bmi_catz'])
df['bp_enc'] = label_encoder.fit_transform(df['blood_pressure'])
print(df.head())
```

### gradient boosting classifier
boosting combines multiple weak models; 
first take avg. as base weak model, then calculate residue which is new target;
combine(add) base with multiple weak models to calculate final model

```python
#gradient boosting classifier
gbc = GradientBoostingClassifier()
model = gbc
model.fit(x_train, y_train)

print(f"Train Accuracy: {gbc.score(x_train, y_train)}")
print(f"Test Accuracy: {gbc.score(x_test, y_test)}")
```

```python
#define feature categories
objective_features = ['age', 'gender', 'height', 'weight', 'BMI']
examination_features = ['ap_hi', 'ap_lo', 'cholesterol', 'gluc']
subjective_features = ['smoke', 'alco', 'active']
```

```python
def evaluate_model(features, model):
    model = model
    model.fit(x_train[features], y_train)
    predictions = model.predict(x_test[features])
    #probability score of positive class(with disease)
    probs = model.predict_proba(x_test[features])[:, 1]
    
    return {
        'accuracy': accuracy_score(y_test, predictions),
        'precision': precision_score(y_test, predictions),
        'recall': recall_score(y_test, predictions),
        'f1': f1_score(y_test, predictions),
        'auc': roc_auc_score(y_test, probs)
    }
```

```python
#function to evaluate model
def print_score(model):
    objective_results = evaluate_model(objective_features, model)
    examination_results = evaluate_model(examination_features, model)
    subjective_results = evaluate_model(subjective_features, model)
    combined_results = evaluate_model(objective_features+examination_features \
                                      +subjective_features, model)
    
    print(f"Objective_results: {objective_results} \n\n, Examination_results: \
          {examination_results} \n\n, Subjective_results: {subjective_results} \
          \n\n, Combined_results: {combined_results}")
```

### ROC curve

```python
#function to get probability scores
def probability_score(features, x_train, x_test, y_train, y_test, model):
    model = rf
    model.fit(x_train[features], y_train)
    #for positive class
    probs = model.predict_proba(x_test[features])[:, 1]
    #calculate roc curve
    fpr, tpr, thresholds = roc_curve(y_test, probs)
    #calculate auc curve
    auc_score = roc_auc_score(y_test, probs)
    
    return fpr, tpr, auc_score
```

```python
combined_features = objective_features+examination_features+subjective_features

#evaluate model (assuming voting clf)
fpr, tpr, auc_score = probability_score(combined_features, x_train, x_test, 
                                       y_train, y_test, rf)
```

```python
#plot roc curve
plt.figure(figsize=(10,6))
plt.plot(fpr, tpr, color='darkorange', label=f'ROC Curve (area = {auc_score:.2f})')
plt.plot([0,1], [0,1], color='red', ls='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```
